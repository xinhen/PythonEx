```python
import networkx as nx
g=nx.DiGraph()
g.add_node(1,name='n1',weight=2)
g.add_node(2,name='n2',weight=2)
g.add_nodes_from([3,4])
g.nodes(data=True)
# g.remove_node(2)
# g.remove_nodes_from([3,4])
g._node[1].update({'name':'n1 new'})
print(g._node)
print(g.node[1])
g.add_weighted_edges_from([(1,2,3.5),(2,1,3.7)])
print(g.edges)
print(g[1][2])
print(g[2][1])
print(g.get_edge_data(1,2))
print(g.has_edge(1,2))

import matplotlib.pyplot as plt
%matplotlib inline
nx.draw(g,pos=nx.spectral_layout(g),nodecolor='r',edge_color='b')
```
